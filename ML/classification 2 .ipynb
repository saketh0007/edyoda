```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
hr_data = pd.read_csv('https://raw.githubusercontent.com/edyoda/data-science-complete-tutorial/master/Data/HR_comma_sep.csv.txt')
hr_data.rename(columns={'sales':'dept'},inplace=True)
```

```python
hr_data.head()
```

```python
hr_data.left.value_counts()
```

### infarences
* Target is descrete in nature so it is a classification problem

```python
target_data = hr_data.left
```

```python
target_data.head()
```

### Taking care of categorical columns
* dept & salary are categorical data
<hr>
* ML Algorithms only understands numeric data
<hr>
* we will use preprocessor LabelEncoder to convert categorical data to numeric data
<hr>

```python
from sklearn.preprocessing import LabelEncoder
```

```python
#le.fit will train the tranisformer
```

```python
le_salary = LabelEncoder()
```

```python
le_salary.fit(hr_data.salary)
```

```python
#le_salary.transform(['low'])
```

```python
#creating new column
hr_data['salary_en'] = le_salary.transform(hr_data.salary)
```

```python
hr_data.head()
```

##### *we can also use fit_transform

```python
#for department
```

```python
le_dept = LabelEncoder()
```

```python
le_dept.fit(hr_data.dept)
```

```python
le_dept.transform(['sales'])
```

```python
hr_data['dept_en'] = le_dept.transform(hr_data.dept)
```

```python
hr_data.head(1)
```

```python
le_dept.classes_
```

```python
#getting feature data from the hr_data
feature_data = hr_data.drop(columns = ['salary','left','dept'])
```

```python
feature_data.columns
```

##### splitting traget data and feature data into test and train data

```python
from sklearn.model_selection import train_test_split
```

```python
trainX,testX,trainY,testY = train_test_split(feature_data,target_data)
```

```python
trainX [:2]
```

###### observation: we are not training a data with string information
<hr>

### Classification Algorithms
*Logistic Regression,RandomForestClassifier,DecisiconTree classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(trainX,trainY)
```

```python
data = [0.38,0.53,2,157,3,1,0,7,2]
```

```python
rf.predict([data])
```

```python
rf.predict(trainX)
```

```python
rf.score(testX,testY)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(trainX,trainY)
```

```python
lr.score(testX,testY)
```

```python
testX['lr_predict'] = lr.predict(testX)
```

```python
testX['rf_predict'] = rf.predict(testX.drop(columns=['lr_predict']))
```

```python
testX['actual'] = testY
```
