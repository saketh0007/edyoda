# Ensemble methods 
* Base Estimators -simple estimators like Linear Regression ,Decision Tree,NearestNeighbours,Naive bayes
* Ensemble Methods - They combine the some of the base estimators & create predictors 
* Decesion Tree was Base estimator and Random Forest was ensemble methods 
* Ensemble methods models into more robust models 

### Types of ensemble methods 
* Bagging methods : Build several estimators independently and average there predictions . Examples : Random Forest 
* Boosting : 
 - Each data has an importent information which is weightage od data 
 - Initially all weightage of the same data 
 - Weightage of the data tells how importent it is to classify  the data correctly 
 - training the model with higher weightage of previously trained model for missclassified data 
 - For prediction all the weak classifiers are consulted

## Voting Classifier 
* As of now same base estimators are used
* How about if we want to combine different types of base estimators 
* Hard Voting : same weightage for all the base estimators 
* soft Voting : Different weightage for the base estimators


```python
from sklearn.ensemble import VotingClassifier
```

```python
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.datasets import load_digits
```

```python
from sklearn.model_selection import train_test_split
```

```python
estimators = [ 
    ('rf',RandomForestClassifier(n_estimators=20)),
    ('svc',SVC(kernel='rbf', probability=True)),
    ('knc',KNeighborsClassifier()),
    ('abc',AdaBoostClassifier(base_estimator=DecisionTreeClassifier() ,n_estimators=20)),
    ('lr',LogisticRegression()) 
]
```

```python
trainX,testX,trainY,testY= train_test_split(digits.data,digits.target)
```

```python
vc = VotingClassifier(estimators=estimators,voting = 'hard')
```

```python
vc.fit(trainX,trainY)
```

```python
vc.score(testX,testY)
```

```python
vc.estimators_
```

```python
vc.estimators
```

```python
for est,name in zip(vc.estimators_,vc.estimators):
    print (name[0], est.score(testX,testY))
```

```python
vc.predict(testX[:2])
```

```python
estimators = [ 
    ('rf',RandomForestClassifier(n_estimators=20)),
    ('knc',KNeighborsClassifier()),
    ('abc',AdaBoostClassifier(base_estimator=DecisionTreeClassifier() ,n_estimators=20)),
    ('lr',LogisticRegression()) 
]
```

```python
vcc = VotingClassifier(estimators=estimators, voting='soft', weights=[1,5,4,1])
```

```python
vcc.fit(trainX,trainY)
```

```python
for est,name in zip(vcc.estimators_,vcc.estimators):
    print (name[0], est.score(testX,testY))
```

```python
vcc.score(testX,testY)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import MinMaxScaler
```

```python
pipeline = make_pipeline(MinMaxScaler(), AdaBoostClassifier(n_estimators=100, base_estimator=DecisionTreeClassifier()))
```
